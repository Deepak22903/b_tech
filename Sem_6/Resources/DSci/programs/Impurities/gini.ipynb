```python
import pandas as pd
import math
from table import *
```

```python
df=pd.read_csv('weather-numerical.csv')
# sort by temperature
df=df.sort_values(by='temperature', ascending=True)
# df.head()
df
```

```python
medians=[]
# Median between every two consecutive values of temperature
for i in range(len(df)-1):
    medians.append((df.iloc[i]['temperature']+df.iloc[i+1]['temperature'])/2)

print('Median between every two consecutive values of temperature')
print(medians)
```

```python
less=[]
more=[]
gini=[]
minGini=math.inf
splitVal = 0

labels=[]
theTemps=[]

for i in medians:
    # Count all yes where temperature is less than and equal to median
    a=[df[df['temperature']<=i]['play'].value_counts().get('yes',0),df[df['temperature']<=i]['play'].value_counts().get('no',0)]
    less.append(a)
    # Count all no where temperature is less than and equal to median if not then 0
    b=[df[df['temperature']>i]['play'].value_counts().get('yes',0),df[df['temperature']>i]['play'].value_counts().get('no',0)]
    more.append(b)

    ans1=round(1-((a[0]/sum(a))**2+(a[1]/sum(a))**2),3)
    ans2=round(1-((b[0]/sum(b))**2+(b[1]/sum(b))**2),3)
    gini.append([ans1,ans2])

    temp=round((sum(a)/14)*ans1+(sum(b)/14)*ans2,3)
    theTemps.append(temp)

    # Find minimum gini index
    if temp<minGini:
        minGini=temp
        splitVal=i

    labels.append(str(i))

# print('Less than or equal to median')
# print(less)
# print('More than median')
# print(more)
# print('Gini index')
# print(gini)


t=make_table(
    rows=[less,more,gini,theTemps],
    labels=labels
)
print(t)

print('Minimum gini index')
print(minGini)
print('Split value')
print(splitVal)
```
